```python
import numpy as np
np.random.seed(1)

import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.random.randint(0, 50, 1000)

# 這次讓 y 與 x 正相關，再增加一些雜訊
y = x + np.random.normal(0, 10, 1000)

# 再次用 numpy 裡的函數來計算相關係數
np.corrcoef(x, y)
```

```python
# 再看看正相關的 x,y 分布
plt.scatter(x, y)
```

```python
# 隨機生成 1000 個介於 0~50 的數 x


# 這次讓 y 與 x 正相關，再增加一些雜訊
x = np.random.randint(0, 50, 1000)

# 這次讓 y 與 x 正相關，再增加一些雜訊
y = np.random.normal(0, 10, 1000)-x
np.corrcoef(x, y)
# 再看看正相關的 x,y 分布
plt.scatter(x, y)
```
